```python
def prime(num, check=2):
    if num % check == 0 : 
        return False
    elif check == num // 2 : 
        return True
    else : 
        return prime(num, check+1)
```

```python
prime(121)
```

**latex**

$$ m = \frac {\sum\limits_{i=0}^N (\bar x - x_i) (\bar y - y_i) }{\sum\limits_{i=0}^N {(\bar x - x_i)^2} } $$

$$ c = \bar y - m \bar x $$

```python
x = height
y = weight
```

```python
x_bar = x.mean()
y_bar = y.mean()
```

```python
x_xbar = x - x_bar
y_ybar = y - y_bar
```

```python
m = ((x_xbar)*(y_ybar)).sum() / ((x_xbar)**2).sum()
```

```python
m
```

```python
c = y_bar - m* x_bar
```

```python
c
```

```python
w = m*h + c
```
