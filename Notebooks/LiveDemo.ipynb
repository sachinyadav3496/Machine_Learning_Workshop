```python
from sklearn.model_selection import train_test_split # split data into train test
from sklearn.linear_model import LinearRegression # model which learns from data
from sklearn.metrics import r2_score  # evaluation accuracy
```

```python
from sklearn import datasets
```

```python
house = datasets.load_boston()
```

##### Housing Price Prediction

Step-1 **Prepare Your Data**

```python
house.feature_names
```

```python
print(house.DESCR)
```

```python
features = pd.DataFrame(house.data, columns=house.feature_names)
features.head()
```

```python
target = pd.Series(house.target)
target.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column_name in features.columns:
    sns.scatterplot(x=features[column_name], y=target)
    plt.show()
```

```python
features.head()
```

```python
target.head()
```

Split data into training and testing

```python
train_X, test_X, train_y, test_y = train_test_split(features, target, 
                                                    test_size=0.3)
```

##### select representation or model

```python
model = LinearRegression()
```

Let's train our model

```python
model.fit(train_X, train_y)
```

```python
m = model.coef_
```

```python
m
```

```python
c  = model.intercept_
c
```

```python
model.predict(features.loc[3:5,]) # pred
```

```python
target.loc[3:5] # actual
```

```python
pred_y  = model.predict(test_X)
```

```python
r2_score(test_y, pred_y )
```
