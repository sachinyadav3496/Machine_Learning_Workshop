```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
city = pd.read_csv("city.csv")
```

```python
city.head()
```

```python
c = ['red', 'green', 'blue', 'cyan', 'magenta', 'yellow', 'orange', 'pink', 'black', 'purple',
    'skyblue', 'gold', 'violet', 'gray', 'indigo']
```

```python
plt.figure(figsize=(10, 5))
city.groupby('country_code')['name'].count().sort_values(ascending=False).head(15).plot(
    kind='bar',color=c, alpha=0.8)

plt.yticks(fontsize=20, color='#123456')
plt.xlabel("Country Code", fontsize=25, color='blue')
plt.ylabel("Frequency", fontsize=25, color='blue')
city.groupby('country_code')['name'].count().sort_values(ascending=False).head(15).plot(
    color='k', marker='>', lw=3, ms=10, alpha=0.5)
plt.xticks(fontsize=20, color='#123456', rotation=90,)
plt.xlim([-1,15])
plt.show()
```
